```python
import anndata as ad
import scanpy as sc
import pandas as pd
```

```python
adataA = ad.read_h5ad('../data/processed_data/k03_80um.h5ad')
adataB = ad.read_h5ad('../data/processed_data/k12_80um_Dpm.h5ad')
```

```python
adataA.var.index = pd.Series(adataA.var.index).astype(float).round(4)
adataB.var.index = pd.Series(adataB.var.index).astype(float).round(4)
adataA.var.index = adataA.var.index.astype(str)
adataB.var.index = adataB.var.index.astype(str)
```
